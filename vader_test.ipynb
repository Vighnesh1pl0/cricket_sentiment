```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

```python
#sentiment_commentry is object
sentiment_commentry = SentimentIntensityAnalyzer()
```

```python
commentary_lines = [
    "Rohit smashes it over mid-on for a massive SIX!",  # positive
    "OUT! Caught behind, big wicket for Mumbai!",        # negative
    "Good length delivery, batsman defends.",            # neutral
    "Pollard smashes a straight six and the ball goes outside the stadium",  # positive
    "It's a dot ball, there are straight 5 dot balls in over",  # negative
    "OUT! Big wicket for Mumbai"   # negative
]
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
sentiments_commentry = SentimentIntensityAnalyzer()

for line in commentary_lines:
    scores = sentiments_commentry.polarity_scores(line)
    print(f"Line: {line}")
    print(f"Compound Score: {scores['compound']}")
    print("---")
```

```python
def cricket_score(line):
    keywords = {
        "SIX"   : +0.5,
        "FOUR"  : +0.3,
        "WICKET": -0.5,
        "OUT"    : -0.5,
        "DOT"   : -0.2,
    }
    score = 0
    for word, value in keywords.items():
        if word in line.upper():  # upper() kyun? soch!
            score += value
    
    return score
```

```python
for line in commentary_lines:
    vader = sentiments_commentry.polarity_scores(line)["compound"]
    cricket = cricket_score(line)
    final = vader + cricket
    print(f"Line: {line}")
    print(f"VADER: {vader} | Cricket: {cricket} | Final: {final}")
    print("---")

if final >= 0.3:
    mood = "🟢 Batting team dominating!"
elif final <= -0.3:
    mood = "🔴 Bowling team on top!"
else:
    mood = "🟡 Even contest"

print(f"Mood: {mood}")
```
